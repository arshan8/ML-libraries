| Feature         | `iloc` (Index-based)                  | `loc` (Label-based)                     |
|----------------|--------------------------------------|-----------------------------------------|
| **Index Type**  | Integer positions (`0, 1, 2`)       | Explicit labels (`'x', 'y', 'z'`)       |
| **Includes End?** | **Excludes** last index (`[0:2]` gives `0,1`) | **Includes** last label (`['x':'y']` gives `'x' and 'y'`) |
| **Column Access** | `df.iloc[:, 0]` (first column)    | `df.loc[:, 'A']` (column A)             |
| **Error Handling** | May raise `IndexError` if out of bounds | May raise `KeyError` if label is missing |

## **When to Use Which?**
✅ **Use `iloc`** when dealing with positions (like `df.iloc[0, 1]`).  
✅ **Use `loc`** when working with meaningful labels (like `df.loc['row_label', 'column_label']`).
