```python
#multilevel indexing + concat

hospital1 = pd.DataFrame({
    "A":[1,2,3],
    "B":[5,7,8],
    "C" : [9,9,9]
})
hospital2 = pd.DataFrame({
    "A":[9,9,3],
    "B":[55,8,111],
    "C" : [9,9,9]
})
hospital3 = pd.DataFrame({
    "A":[31,32,3],
    "B":[53,37,8],
    "C" : [93,39,9]
})

frames = [hospital1,hospital2,hospital3]

pd.concat(frames, axis = 0)
```

```python
pd.concat(frames, axis = 0,ignore_index=True)
```

```python
#multilevel index

g =pd.concat(frames, axis = 0,keys = ["hospital 1","hospital2","hospital3"])  #never include ignore index command here please
```

```python
pd.concat(frames, axis = 1,keys = ["Dtype 1","dType 2","d 3"]) 
```

```python
#merge: performs join like operaton

data1 = pd.DataFrame({
    "keys":["k1","k2","k3"],
    "B":[5,7,8],
    "C" : [9,9,9]
})
data2 = pd.DataFrame({
    "A":[9,9,3],
   "keys":["k1","k2","k3"],
    "C" : [9,9,9]
})

print("\n",data1, "\n \n", data2,"\n")
```

```python
pd.merge(data1,data2, on = "keys")
```
